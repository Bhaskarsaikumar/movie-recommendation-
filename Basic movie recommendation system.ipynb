```python
import pandas as pd
import numpy as np
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits.head()
```

```python
movies.head()
```

```python
print('credits:',credits.shape)
print('movies:',movies.shape)
```

```python
credits_column_rename=credits.rename(index=str,columns={'movie_id':'id'})
movies_df_merge=movies.merge(credits_column_rename,on='id')
movies_df_merge.head()
```

```python
movies_cleaned_df=movies_df_merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
movies_cleaned_df.head()
```

```python
movies_cleaned_df.info()
```

# weighted average formula 


# W= Rv+Cm/v+m

# w=weighted rating
# R=average for the movie as anumber from 0 to 10(mean)=Rating
# v =number of votes to movie
# C=mean of the vote across for whole report
# m=min votes required to make in list top 250

```python
#will calculate based on forumla
v=movies_cleaned_df['vote_count']
R=movies_cleaned_df['vote_average']
C=movies_cleaned_df['vote_average'].mean()
m=movies_cleaned_df['vote_count'].quantile(0.70)
```

```python
movies_cleaned_df['weighted_average']=((R*v)+(C*m))/(v+m)
```

```python
movies_cleaned_df.head()
```

```python
movie_sorted_rankings=movies_cleaned_df.sort_values('weighted_average',ascending=False)
movie_sorted_rankings[['original_title','vote_count','vote_average','weighted_average','popularity']].head(15)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

weight_average=movie_sorted_rankings.sort_values('weighted_average',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weighted_average'].head(10),y=weight_average['original_title'].head(10))
plt.xlim(4,10)
plt.title('best movies based on movies average votes',weight='bold')
plt.xlabel('weighted average score',weight='bold')
plt.ylabel('movie title',weight='bold')
plt.savefig('best_movies.png')
```

```python

popularity=movie_sorted_rankings.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
ax=sns.barplot(x=popularity['popularity'].head(10),y=popularity['original_title'].head(10),data=popularity)

plt.title('popular by votes',weight='bold')
plt.xlabel('popularity score',weight='bold')
plt.ylabel('movie title',weight='bold')
plt.savefig('popular_movies.png')
```

# recommendation based on weighted average and popularity(priority giving 50% both)

```python
from sklearn.preprocessing import MinMaxScaler
scaling=MinMaxScaler()
movie_scaled_df=scaling.fit_transform(movies_cleaned_df[['weighted_average','popularity']])
movies_normalized_df=pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity'])
movies_normalized_df.head()
```

```python
movies_cleaned_df[['normalized_weight_average','normalized_popularity']]=movies_normalized_df
```

```python
movies_cleaned_df.head()
```

```python
movies_cleaned_df['score']=movies_cleaned_df['normalized_weight_average']*0.5+movies_cleaned_df['normalized_popularity']
movies_scored_df=movies_cleaned_df.sort_values(['score'],ascending=False)
movies_scored_df[['original_title','normalized_weight_average','normalized_popularity','score']].head()
```

```python

scored_df=movies_cleaned_df.sort_values('score',ascending=False)
plt.figure(figsize=(16,6))
ax=sns.barplot(x=scored_df['score'].head(10),y=scored_df['original_title'].head(10),data=scored_df)

plt.title('best rated & most popular',weight='bold')
plt.xlabel('score',weight='bold')
plt.ylabel('movie title',weight='bold')
plt.savefig('score_movies.png')
```
